```python
%cd /m2docs/res/
# roboflow option:

# Export code snippet and paste here
#!curl -L "https://app.roboflow.ai/ds/f2ILNFXcex?key=GIMpMcsbeE" > roboflow.zip; unzip roboflow.zip; rm roboflow.zip
#!curl -L "https://app.roboflow.ai/ds/h6LDmZVC32?key=JYVDlRLhyf" > roboflow.zip; unzip -o roboflow.zip; rm roboflow.zip
#!curl -L "https://app.roboflow.ai/ds/GXCFEC59RW?key=nzvPu7wgN2" > roboflow.zip; unzip -o roboflow.zip; rm roboflow.zip
#!curl -L "https://app.roboflow.ai/ds/HzuJNJlYQm?key=Ma4hIDVWMl" > roboflow.zip; unzip -o roboflow.zip; rm roboflow.zip


# local files option:
import os
import glob
import cv2
from os import listdir
from os.path import isfile, join
from shutil import copyfile

# split num:
test_count = 200
val_count = 400

# Resize images?
# This will make the program take MUCH longer to run, usually
resize = True
length = 512

# change to most recent directory:
src_dir = "/m2docs/res/label_data_2020819_15:10"
source_files = glob.glob(src_dir + "/*.jpg")

print(source_files)

# directory locations:
%mkdir /m2docs/res/train
%mkdir /m2docs/res/test
%mkdir /m2docs/res/valid
%mkdir /m2docs/res/train/images
%mkdir /m2docs/res/train/labels
%mkdir /m2docs/res/test/images
%mkdir /m2docs/res/test/labels
%mkdir /m2docs/res/valid/images
%mkdir /m2docs/res/valid/labels
train_dir = "/m2docs/res/train"
test_dir = "/m2docs/res/test"
val_dir = "/m2docs/res/valid"

count = 0
for file in source_files:
    if count < test_count:
        if resize:
            try:
                im_arr = cv2.imread(file, cv2.IMREAD_COLOR)
                im_resized = cv2.resize(im_arr, (length, length))
                cv2.imwrite(os.path.join(test_dir,'images/',file.replace(src_dir+"/","")), im_resized)
            except:
                print("error with img.")
        else:
            copyfile(file, os.path.join(test_dir,'images/',file.replace(src_dir+"/","")))
        copyfile(file.replace('.jpg','.txt'), os.path.join(test_dir,'labels/',file.replace(src_dir+"/","").replace('.jpg','.txt')))
    elif count < val_count:
        if resize:
            try:
                im_arr = cv2.imread(file, cv2.IMREAD_COLOR)
                im_resized = cv2.resize(im_arr, (length, length))
                cv2.imwrite(os.path.join(val_dir,'images/',file.replace(src_dir+"/","")), im_resized)
            except:
                print("error with img.")
        else:
            copyfile(os.path.join(src_dir,file), os.path.join(val_dir,'images/',file.replace(src_dir+"/","")))
        copyfile(file.replace('.jpg','.txt'), os.path.join(val_dir,'labels/',file.replace(src_dir+"/","").replace('.jpg','.txt')))
    else:
        if resize:
            try:
                im_arr = cv2.imread(file, cv2.IMREAD_COLOR)
                im_resized = cv2.resize(im_arr, (length, length))
                cv2.imwrite(os.path.join(train_dir,'images/',file.replace(src_dir+"/","")), im_resized)
            except:
                print("error with img.")
        else:
            copyfile(os.path.join(src_dir,file), os.path.join(train_dir,'images/',file.replace(src_dir+"/","")))
        copyfile(file.replace('.jpg','.txt'), os.path.join(train_dir,'labels/',file.replace(src_dir+"/","").replace('.jpg','.txt')))
    count+= 1
    
print("done.")
```

## Cropping
* Copy prediction `*.txt` files to a new directory
* Threshold predictions based on size of detected region
* Crop images and save in new directory

```python
%rm /m2docs/res/predicted_text/*
%cp /m2docs/res/yolov5/inference/output/*.txt /m2docs/res/predicted_text 
```

```python
#%rm /m2docs/res/cropped_imgs/*
from os import listdir
from os.path import isfile, join
import cv2

###	thresholding parameters
min_size = 100	# minimum size for detection, in pixels

### output parameters
square = True	# square the cropped image
padding = 0 # number of pixels to pad around the image
IMG_DIR = '/m2docs/res/bees_unsorted_all'	# images directory
TXT_DIR = '/m2docs/res/predicted_text/'	  # text   directory
OUT_DIR = '/m2docs/res/cropped_imgs'	# output directory

text_files = listdir(TXT_DIR)
img_files = listdir(IMG_DIR)

#print(text_files)
#print(img_files)

error_count = 0
thresh_count = 0
complete_count = 0
skip_count = 0

# for each inference file:
for file in text_files:
    if not isfile(join(OUT_DIR,file.replace('.txt','.jpg'))):
        print("\n")
        print("Reading inference",file)
        # open inference and image
        inference = open(join(TXT_DIR,file),"r")
        image = cv2.imread(join(IMG_DIR,file.replace('.txt','.jpg')))

        # get image dims
        [img_dim_x, img_dim_y, z] = image.shape
        print(img_dim_x, img_dim_y, z)

        for line in inference:
            # read inference line. Formatted as <0> <x> <y> <xlen> <ylen>
            [id, x, y, xlen, ylen] = [float(i) for i in line.split()]

            box_center_x = int(y*img_dim_x)
            box_center_y = int(x*img_dim_y)
            box_length	 = int(xlen*img_dim_x)
            box_height	 = int(ylen*img_dim_y)

            # debug prints: print info about the file
            print(id, x, y, xlen, ylen)
            print("-> Centered at", box_center_x, box_center_y)
            print("-> The box dimensions are", box_length, box_height)

            if padding > 0:
                box_length += padding
                box_height += padding
                print("-> Padded to",box_length,box_height)

            # threshold image: check if the detected box is too small
            if (box_length < min_size or box_height < min_size):
                # not quite big enough
                print("fails size thresholding")
                thresh_count += 1
            else:
                edge_length = 0
                if (square):
                    edge_length = max(box_length, box_height)
                    (x1, x2, y1, y2) = ((box_center_x - int(edge_length/2)),
                                        (box_center_x + int(edge_length/2)), 
                                        (box_center_y - int(edge_length/2)),
                                        (box_center_y + int(edge_length/2)))
                    print("Bounds:",(x1, x2, y1, y2))
                else:
                    (x1, x2, y1, y2) = ((box_center_x - int(box_length/2)),
                                        (box_center_x + int(box_length/2)), 
                                        (box_center_y - int(box_height/2)),
                                        (box_center_y + int(box_height/2)))
                    print("Bounds:",(x1, x2, y1, y2))

                try:
                    # head off out of bounds errors
                    if x1 < 0: x1 = 0
                    if y1 < 0: y1 = 0
                    if x2 > img_dim_x: x2 = img_dim_x
                    if y2 > img_dim_y: y2 = img_dim_y

                    print("Cropping to:",(x1, x2, y1, y2))
                    cropped_image = image[x1:x2, y1:y2]

                    #save cropped image
                    cv2.imwrite(join(OUT_DIR,file.replace('.txt','.jpg')), cropped_image)
                    complete_count += 1
                except:
                    print("error occured while writing image")
                    error_count += 1
    else:
        print("Skipped")
        skip_count += 1

print("")
print("Process completed. Errors:", error_count)
print("   Images thresholded out:", thresh_count)
print("          Total completed:", complete_count)
print("                  Skipped:", skip_count)
```

# Copy files to new folder

Next we copy the prediction texts and the images to a new folder. This data should then be added to the initial training set, after briefly checking that the predictions are _all_ correct. By setting the threshold high in the detection step, we can be sure that the weights adjust off of these good predictions.

```python
#timestamp new dataset
import datetime
import os
from shutil import copyfile

IMG_DIR = '/m2docs/res/bees_unsorted_all'	# images directory
TXT_DIR = '/m2docs/res/predicted_text'	  # text   directory
OUT_DIR = '/m2docs/res/cropped_imgs'	# output directory
SKIP_FILE = '/m2docs/res/skip_list.txt'
skiplist = []

try:
    with open(SKIP_FILE) as file:
        for line in file:
            skiplist.append(line.strip())
except:
    print("[w] Skip file does not exist.")
    print("[w] If this is NOT the first run, something has gone wrong.")
    
current_time = datetime.datetime.now()
dir_name = 'label_data_' + str(current_time.year) + str(current_time.month) + str(current_time.day) + '_' +str(current_time.hour) + ':' +str(current_time.minute) + '/'

if not os.path.exists(os.path.join('/m2docs/res',dir_name)):
    os.makedirs(os.path.join('/m2docs/res',dir_name))
    print("[ ] Created " + os.path.join('/m2docs/res',dir_name))
else:
    print("[!] Directory already exists")
    
print("[ ] Copying images and textfiles...")
for fileName in os.listdir(OUT_DIR):
    if fileName in skiplist:
        print("[a]",fileName,"already in set, skipping")
    else:
        try:
            copyfile(os.path.join(IMG_DIR,fileName), os.path.join('/m2docs/res',dir_name,fileName))
        except:
            print('[!] Error encountered when copying image', fileName)
        try:
            copyfile(os.path.join(TXT_DIR,fileName.replace('.jpg','.txt')), os.path.join('/m2docs/res',dir_name,fileName.replace('.jpg','.txt')))
        except:
            print('[!] Error encountered when copying text', fileName)
print('[.] Done copying')
```
